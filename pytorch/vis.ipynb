```python
from model import net16
from torchvision.models import resnet50, mobilenet_v2
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchviz import make_dot, make_dot_from_trace


from torchsummary import summary



model_1 = resnet50(pretrained=True)
num_ftrs = model_1.fc.in_features
model_1.fc = nn.Linear(num_ftrs, 153)

model_2 = mobilenet_v2(pretrained=True)
model_2.features[0][0] = nn.Conv2d(6, 32, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False)
fc_features = model_2.classifier[1].in_features
model_2.classifier[1] = nn.Sequential(
                nn.Linear(fc_features, 1),
                nn.Sigmoid())
```

```python
summary(model_1, (3,224, 224))
```

```python
summary(model_2, (6,224,224))
```

```python
from torchviz import make_dot, make_dot_from_trace

x = torch.randn(1, 3, 224, 224).requires_grad_(True)
y = net16(x)
dot = make_dot(y, params=dict(list(net16.named_parameters()) + [('x', x)]))
dot.render(format='png',filename='vgg16')
```

```python
x = torch.randn(1, 3, 224, 224).requires_grad_(True)
y = model_1(x)
dot = make_dot(y, params=dict(list(model_1.named_parameters()) + [('x', x)]))
dot.render(format='png',filename='pretrained')
```

```python
x = torch.randn(1, 6, 224, 224).requires_grad_(True)
y = model_2(x)
dot = make_dot(y, params=dict(list(model_2.named_parameters()) + [('x', x)]))
dot.render(format='png',filename='compatiblity')
```
